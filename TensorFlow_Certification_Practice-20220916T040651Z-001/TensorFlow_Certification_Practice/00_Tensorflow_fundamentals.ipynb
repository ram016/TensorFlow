### Matrix Multiplication with tensor

**In machine learning, matrix multiplication  is one of the most common  tensor operations.**

**There are two rules  our tensor(or matrices) need to  fulfil  if we  going  to matrix multiply  them:-** 
 1. The inner dimension must be matches.
 2. The resulting  matrix  has the  shape of the outer  dimensions.

```python
tensor

```

```python
tf.matmul(tensor,tensor)
```

```python
tensor,tensor
```

```python
tensor*tensor # multiply each item with specific index.
```

### Aggregating tensors
  * **Aggregate tensrs = condensing them from the multiple values down to a ** **smaller  amount of values**.

```python
  # Get the absolute values

tns=tf.constant([-7,-10])
print("Original tensor",tns)
print("Absolute Value of tns tensor ",tf.abs(tns))
```

###Let's go through  the following  forms of Aggregation 
  * Get the minimum
  * Get the maximum
  * Get the mean of a tensor
  * Get the sum of a tensor

```python
# Create a random tensor with the value between 0 and 100 of size 50

ag = tf.constant(np.random.randint(1,100,size=50))
ag
```

```python
# getting minimum value
tf.reduce_min(ag)
```

```python
# getting maximum value
tf.reduce_max(ag)
```

```python
# get mean value

tf.reduce_mean(ag)
```

```python
#get sum of the tensor
tf.reduce_sum(ag)
```

```python
# Variance of the tesor.

import tensorflow_probability as tfp
tfp.stats.variance(tf.cast(ag, dtype=tf.float32))
```

```python
# tf.cast(ag,dtype=tf.float32)
```

```python
#standard deviation of tensor
x = tf.constant([[1., 2.], [3., 4.]])
tf.math.reduce_std(x)
```

```python
tfp.stats.stddev(x, sample_axis=0, keepdims=False, name=None)
```
