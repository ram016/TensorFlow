# **To visualize our model's  predictions,let's create a function plot_decission_boundary(),this function will:**

 * Take in trained model,features(X) and labels(y).
 * Create meshgrid of the different X Values.
 * Make prediction across meshgrid.
 * plot the predictions as well as line between zones(where each unique class falls)

```python
import numpy as np
```

```python
def plot_decission_boundary(model=model_2,X=X,y=y):
    """
    Plot the decission boundary created by model predicting on X.
    This function taken from 2Resource which is:-
      1.ðŸ—¡ https://cs231n.github.io/neural-networks-case-study/
      2.ðŸ—¡ Resource: https://github.com/davydw/madewithML-basics
    """

    # Define the axis boundaries of the plot and create a meshgrid.
    x_min,x_max = X[:,0].min()-0.1 , X[:,0].max()+0.1
    y_min,y_max =  X[:,1].min()-0.1 , X[:,1].max()+0.1

    xx,yy = np.meshgrid(np.linspace(x_min,x_max,100),
                        np.linspace(y_min,y_max,100))

    # Create X value (we are ging to prediction on these)
    x_in = np.c_[xx.ravel() , yy.ravel()] # stack 2D arrays together.

    # make predictions
    y_pred = model.predict(x_in)

    # check for multiclass
    if len(y_pred[0])>1:
      print("Doing multiclass classification")
      # we have to reshape our prediction to get them redy for plotting 
      y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
    else:
      print("Doing binary classification ")
      y_pred = np.round(y_pred).reshape(xx.shape)

    # plot the decission boundary.
    plt.contourf(xx,yy,y_pred,cmap=plt.cm.RdYlBu,alpha=0.7)
    plt.scatter(X[:,0],X[:,1],c=y,s=40 ,cmap=plt.cm.RdYlBu)
    plt.xlim(xx.max(),xx.min())
    plt.ylim(yy.max(),yy.min())

```

```python
plot_decission_boundary(model=model_2,X=X,y=y)
```

# **Finding the best learning rate**
***(the learning rate where the loss decreases most** **during the training ) we're going to use the following steps:-**
   * A learning rate **callback** -callback as extra piece of functionality we can add to your **while** its training.
   * Create another model 
   * A **modified** the loss curve plot

```python
 # Create the new model

 # set random seed
tf.random.set_seed(42)

 # create a model
model_9 = tf.keras.Sequential([
            tf.keras.layers.Dense(4,activation="relu"),
            tf.keras.layers.Dense(4,activation="relu"),
            tf.keras.layers.Dense(1,activation="sigmoid")
 ])

 #Compile the model
model_9.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                  optimizer="Adam",
                  metrics=["accuracy"])
 
 # create a learning rate callback 
lr_schedular = tf.keras.callbacks.LearningRateScheduler(lambda epochs: 1e-4 * 10**(epochs/20))

# Fit the model

history_9 = model_9.fit(X_train,
            y_train,
            epochs=100,
            callbacks=[lr_schedular])
```

# Building the multi-class classification model

**For our multi-class classification model,we can use a similar architecture to our binary classifiers,however we're going to have to tweaka few things:**
 * Input shape =28x28(Shape of one image).
 * Output shape = 10(one per class of clothing).
 * loss function = tf.keras.losses.CategoricalCrossentrophy().
     * **if our labels are one-hot encoding,use CategoricalCrossentrophy().**
     * **if our labels are in integer form then, use SparseCategoricalCrossentrophy()**.
 * Output layer activation = Softmax(not sigmoid).

```python
# Set the random seed
tf.random.set_seed(42)

# Crete model
model_11 = tf.keras.Sequential([
            tf.keras.layers.Flatten(input_shape=(28,28)),
            tf.keras.layers.Dense(4,activation="relu"),
            tf.keras.layers.Dense(4,activation="relu"),
            tf.keras.layers.Dense(10,activation=tf.keras.activations.softmax)
])

# compile the model
model_11.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
                 optimizer=tf.keras.optimizers.Adam(),
                 metrics=["accuracy"])

# Fit the model
non_norm_history = model_11.fit(train_data,
                                train_labels,
                                epochs=10,
                                validation_data=(test_data,test_labels))
```

**Neural netwok prefer data to be scaled(or normalized)this means they like to have the numbers in the tensor they try to find the pattern between 0 and 1.**

```python
# we can get our training data and testing data between 0 & 1 by dividingby the max value of train_data/.

train_data_norms = train_data/255.0
test_data_norms = test_data/255.0

# the min and max of normalized data.
train_data_norms.min() , test_data_norms.max()
```

```python
# Now our data get Normalized ,let's build a model to find pattern in it.

# set the random seed
tf.random.set_seed(42)
# create the model

model_12 = tf.keras.Sequential([
            tf.keras.layers.Flatten(input_shape=(28,28)),
            tf.keras.layers.Dense(4,activation="relu"),
            tf.keras.layers.Dense(4,activation="relu"),
            tf.keras.layers.Dense(10,activation=tf.keras.activations.softmax)
])

# compile the model

model_12.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
                 optimizer=tf.keras.optimizers.Adam(),
                 metrics=["accuracy"])

# Fit the model
norm_history = model_12.fit(train_data_norms,
                            train_labels,
                            epochs=10,
                            validation_data=(test_data_norms,test_labels))
```

# Finding the Ideal learning Rate

```python
# set the random seed
tf.random.set_seed(42)
# create the model

model_13 = tf.keras.Sequential([
            tf.keras.layers.Flatten(input_shape=(28,28)),
            tf.keras.layers.Dense(4,activation="relu"),
            tf.keras.layers.Dense(4,activation="relu"),
            tf.keras.layers.Dense(10,activation=tf.keras.activations.softmax)
])

# compile the model

model_13.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(),
                 optimizer=tf.keras.optimizers.Adam(),
                 metrics=["accuracy"])

# create learning rate callback
lr_scheduler = tf.keras.callbacks.LearningRateScheduler(lambda epochs: 1e-3 * 10**(epochs/20))

# Fit the model.


find_lr_history = model_13.fit(train_data_norms,
            train_labels,
            epochs=40,
            validation_data=(test_data_norms,test_labels),
            callbacks=[lr_scheduler])
```

```python
model_13.evaluate(train_data_norms)
```

```python
from IPython.core.pylabtools import figsize
import itertools
from sklearn.metrics import confusion_matrix

def confusion_matrix(y_true,y_pred,classes,figsize=(10,10),text_size=15):
  
  # create confusion matrix
  cm =  confusion_matrix(y_true,y_preds)  

  # normalize the consusion matrix
  cm_norm = cm/cm.sum(axis=1)[:,np.newaxis]
  n_classes = cm.shape[0]

  # let's pretty it
  fig,ax = plt.subplots(figsize=figsize)
  # matrix plot
  cax =  ax.matshow(cm,cmap=plt.cm.RdYlBu)
  fig.colorbar(cax)

  # set labels to the classes
  if classes:
    labels=classes
  else:
    labels=np.arange(cm.shape[0])


  # label the axes
  ax.set(title="Confusion matrix",
        xlabel="Prediction Label",
        ylabel="True Label",
        xticks=np.arange(n_classes),
        yticks=np.arange(n_classes),
        xticklabels=labels,
        yticklabels=labels)

  # set x-axis label to bottom 
  ax.xaxis.set_label_position("bottom")
  ax.xaxis.tick_bottom()

  # Adjust label size
  ax.title.set_size(text_size)
  ax.xaxis.label.set_size(text_size)
  ax.yaxis.label.set_size(text_size)

  # set the threshold for different colors
  threshold = (cm.max()+cm.min())/2

  # plot the text on each cell
  for i,j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
    plt.text(j,i,f"{cm[i,j]},({cm_norm[j,i]*100:.1f}%",
              horizontalalignment="center",
              color="red" if cm[i,j]>threshold else "black",
              size=text_size)



```

**Note:-** Often when working with images and other form of visual data,it's a good idea to vizualize as developa further understanding of the data and the inputs and outputs of the model.

**How about we create a fun little function for:**
 * PLot random image
 * Make prediction on said image.
 * Label the plot with the truth label and the prediction label

```python
import random

def plot_random_image(model,images,true_label,classes):
  """
  Picks a random images,plot it and labels it with a prediction and truth labels
  """

  #set the random integer
  i = random.randint(0,len(images))

  # create predictin and target
  target_image = images[i]
  pred_probs = model.predict(target_image.reshape(1,28,28)) 
  pred_label = classes[pred_probs.argmax()]
  true_label = classes[true_label[i]]

  #plot images.
  plt.imshow(target_image,cmap=plt.cm.binary)
  
  if pred_label== true_label:
    color="green"
  else: 
    color="red"

   # Add xlabel information (prediction/true_labels)
  plt.xlabel("Pred: {} {:2.0f}% (True: {})".format(pred_label,
                                                    100*tf.reduce_max(pred_probs),
                                                    true_label,
                                                    color= color))
```
